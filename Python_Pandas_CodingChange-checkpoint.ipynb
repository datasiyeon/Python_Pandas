```python
def edu_groups(series):  #edu_groups 정의
    if series==2:
        return 1
    elif series==3 or series==4:
        return 2
    elif series >=5:
        return 3
```

```python
df['eduM']=df['edu'].apply(edu_groups) #edu_groups로 변경 (apply함수)
df['eduM'].value_counts(sort=False)
```
